```python
import pandas as pd, ast

tinfo = pd.read_csv("/home/zburnside/pga_prediction_model/tournament_data.csv")

tinfo["course_name"] = tinfo["courseData"].apply(
    lambda x: (ast.literal_eval(x).get("name") if pd.notna(x) else None)
)

# One row per tournament name -> course (usually stable)
tournament_to_course = (
    tinfo[["name","course_name"]]
    .dropna()
    .drop_duplicates()
    .set_index("name")["course_name"]
    .to_dict()
)

tournament_to_course
```
